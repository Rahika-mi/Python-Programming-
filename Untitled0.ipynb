```python
# Import packages
import pandas as pd
import numpy as np
from collections import Counter

# Load and clean data
df = pd.read_csv('/content/International_T20_Data.csv')

# Rename columns
df = df.rename(columns={
    'meta.created': 'created_date',
    'info.venue': 'venue',
    'info.city': 'city',
    'info.dates': 'match_date',
    'info.teams': 'teams',
    'info.toss.winner': 'toss_winner',
    'info.toss.decision': 'toss_decision',
    'info.outcome.winner': 'match_winner',
    'info.player_of_match': 'player_of_match'
})

df.columns = df.columns.str.replace('.', '_')

# Analyze team pairings
df['team_pair'] = df['teams'].apply(lambda x: tuple(sorted(x)))
top_pair = df['team_pair'].value_counts().head(1)
print("Most Common Team Pairing:")
print(top_pair)

# Calculate win percentages
matches_played = Counter()

for teams in df['teams']:
    for team in teams:
        matches_played[team] += 1

# Calculate wins (outside the loop)
wins = df['match_winner'].value_counts().to_dict()

win_percentage = {}
for team in matches_played:
    win_percentage[team] = (wins.get(team, 0) / matches_played[team]) * 100

# Display top 5 teams by win percentage
win_df = pd.DataFrame(list(win_percentage.items()), columns=['Team', 'Win_Percentage'])
top5 = win_df.sort_values(by='Win_Percentage', ascending=False).head(5)

print("\nTop 5 Teams by Win Percentage:")
print(top5)

# Function to extract scorecard data
def get_scorecard(innings_df, innings_no):
    """
    Extract top batsmen and bowlers for a specific innings

    Parameters:
    - innings_df: DataFrame with innings data
    - innings_no: Innings number to analyze

    Returns:
    - batsman_score: Top 4 batsmen by runs
    - bowler_wickets: Top 4 bowlers by wickets
    """
    innings_data = innings_df[innings_df['innings'] == innings_no]

    batting_team = innings_data['batting_team'].iloc[0]
    bowling_team = innings_data['bowling_team'].iloc[0]

    # Top 4 batsmen
    batsman_score = (
        innings_data
        .groupby('batsman')['runs_batsman']
        .sum()
        .sort_values(ascending=False)
        .head(4)
        .reset_index()
    )

    # Top 4 bowlers (by wickets)
    bowler_wickets = (
        innings_data[innings_data['player_out'].notna()]
        .groupby('bowler')
        .size()
        .sort_values(ascending=False)
        .head(4)
        .reset_index(name='wickets')
    )

    return batsman_score, bowler_wickets
```
